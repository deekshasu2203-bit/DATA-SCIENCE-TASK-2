```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the latest data from GitHub
train_url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%202/train.csv"
test_url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%202/test.csv"
gender_url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%202/gender_submission.csv"

train_df = pd.read_csv(train_url)
test_df = pd.read_csv(test_url)
gender_df = pd.read_csv(gender_url)

```

```python
# Basic info
print("Train Data Overview:")
print(train_df.head())

```

```python
# Survival count
plt.figure(figsize=(6,4))
sns.countplot(x='Survived', data=train_df, palette='Set2')
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

```python

# Survival by gender
plt.figure(figsize=(6,4))
sns.countplot(x='Sex', hue='Survived', data=train_df, palette='Set1')
plt.title('Survival by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Age distribution
plt.figure(figsize=(8,5))
sns.histplot(train_df['Age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
# Survival by passenger class
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', hue='Survived', data=train_df, palette='Set3')
plt.title('Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```
